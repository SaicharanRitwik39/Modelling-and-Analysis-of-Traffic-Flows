```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def traffic(x,t): 
 a1 = 0 
 a2 = 0 
 a3 = 0 
 a4 = 0 
 a5 = 0 
 
 A = x[0] 
 B = x[1] 
 C = x[2] 
 dAdt = -a1 * A * B + a2 * C + a3 - a4 * A 
 dBdt = a1 * A * B - a5 * B - a4 * B 
 dCdt = a5 * B - a4 * C - a2 * C 
 
 return [dAdt, dBdt, dCdt] 
 
x0 = [80,0,0] #Initial conditions. 
 
t = np.linspace(0,15,1000) 
x = odeint(traffic,x0,t) 
A = x[:,0] 
B = x[:,1] 
C = x[:,2] 
 

plt.xlabel("Time(t)") 
plt.ylabel("F(t)") 

 
plt.plot(t,A) 
plt.show() 
```
